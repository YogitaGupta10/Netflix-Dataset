```python
df['director'].replace(np.nan, 'No Director',inplace=True)
df['country'].replace(np.nan, 'No Country',inplace=True)
df['cast'].replace(np.nan, 'No Cast',inplace=True)
df.isnull().sum()
```

```python
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
```

```python
director_by_country = {}
count_by_country = {}
for i in range(0,len(df['country'])):
    ctry= df['country'].iloc[i].strip()
    if ctry in count_by_country:
        count_by_country[ctry]+=1
    else:
        count_by_country[ctry]=1


    # print(ctry)
    if ctry in director_by_country:
        dr= df['director'].iloc[i]
        if dr in director_by_country[ctry]:
            director_by_country[ctry][dr]+=1
        else:
            director_by_country[ctry][dr]=1

    else:
        dr= df['director'].iloc[i]
        director_by_country[ctry]={dr:1}



print(director_by_country)
print(count_by_country)
```

```python
rows = []
for country, directors in director_by_country.items():
    for director, count in directors.items():
        rows.append({'Country': country, 'Director': director, 'Count': count})

df1 = pd.DataFrame(rows)
print(df1)
```

```python
rows1 = []
for country, count in count_by_country.items():
    rows1.append({'Country': country,'Count': count})

df2 = pd.DataFrame(rows1)
print(df2)
```
